## Singular value decomposition

```julia
m, n = 8,4
A = randn(m,n)
rank(A)
```

```julia
U, s, V = svd(A)
```

```julia
Σ = diagm(s)
```

```julia
A[:,n] = A[:,1:n-1]*randn(n-1)
rank(A)
```

```julia
import Images, ImageView, Colors, TestImages
using Interact
```

```julia
img = TestImages.testimage("mandrill")
img = convert(Images.Image{Colors.Gray}, img)
```

```julia
A = Images.data(img)
U,s,V = svd(A);
```

```julia
using Plots
plot(s,yscale=:log10)
```

```julia
k = 10
B = U[:,1:k] * diagm(s[1:k]) * V[:,1:k]'
B -= min(minimum(B))
Images.grayim(B)
```

```julia
@manipulate for k=slider(0:100)
    B = U[:,1:k] * diagm(s[1:k]) * V[:,1:k]'
    B -= min(minimum(B))
    Images.grayim(B)
end
```
