# Monday, Nov 7
Jacobi and Gauss-Seidel

```julia
using Plots
```

```julia
srand(0)
n = 100
A = randn(n,n); A = A'*A + 1000*eye(n)
b = randn(n);
```

```julia
function matvec(z)
    return A*z
end;
```

```julia
D = diagm(diag(A))
x = zeros(n)
rnorm = []
for i = 1:20
    r = b - matvec(x)
    x = x + D\r
    push!(rnorm, norm(r))
end;
```

```julia
plot(1:20, rnorm, yscale=:log10)
```

```julia
E = tril(A)
x = zeros(n)
rnorm = []
for i = 1:20
    r = b - A*x
    x = x + E\r
    push!(rnorm, norm(r))
end;
```

```julia
plot(1:20, rnorm, yscale=:log10)
```
