# Linear least squares example
Finding $n$-degree polynomial approximation to $cos(2\pi t)$ for $n = 1,\ldots,5$

```julia
function lsfit(t, b, n)

#     function coefs = lsfit (t, b, n)

#     Construct coefficients of the polynomial of
#     degree at most n-1 that best fits data (t,b)

    m = length(t);

#     long and skinny A
    A = ones(m,n);
    for j=1:n-1
        A[:,j+1] = A[:,j].*t;
    end

#     normal equations and solution
    B = A'*A; y = A'*b;
    coefs = B \ y;
    return coefs;
end

m = 21;
tt = [0:1/(m-1):1];
bb = cos(2*pi*tt);

# find polynomial coefficients
coefs = Array{Float64}[]
for n=1:5
    push!(coefs, lsfit(tt,bb,n));
end

# Evaluate and plot
t = [0:.01:1];
t = reshape(t, 101, 1);
z = ones(101,5);
for n=1:5
    for j=1:101
        z[j,n] = polyval(Poly(coefs[n]), t[j])
    end
    j = 1;
end
println(size(t), size(z))
plot(t,z[:,1],"k-",t,z[:,2],"r",t,z[:,3],"b",t,z[:,4],"m",t,z[:,5],"y",tt,bb,"c")
xlabel("t")
ylabel("p_{n-1}")
```

# Least squares solving example

```julia
tqr = zeros(8,1)
tne = zeros(8,1)
for n = 300:100:1000
#   fill a rectangular matrix A and a vector b with random numbers
#   hoping that A'*A is nonsingular
    m = n+1; # or m = 3*n+1 or something else
    A = randn(m,n); b = randn(m,1);
    
#   solve and find execution times. First, Julia way using QR
    tic()
    CPUtic();
    xqr = A \ b;
    tqr[n/100-2] = toq() - CPUtoq(); 
    
#   next use normal equations
    tic()
    CPUtic();
    B = A'*A; y = A'*b;
    xne = B \ y;
    tne[n/100-2] = toq() - CPUtoq(); 
end

ratio = tqr./tne;
plot(300:100:1000,ratio)
xlabel("n")
ylabel("CPU time ratio")
```

# Data missfit 

```julia

t = [1:1:25];
b = [1.0,1.1,1.0,1.2,1.2,1.3,1.4,1.5,1.3,1.4,1.6,1.9,1.6,1.8,1.8,1.9,2.0,2.0,2.1,2.2,2.3,2.2,2.4,2.2,2.5];

coefs = lsfit(t,b,2);

v = coefs[1] + coefs[2]*t;

plot(t,b,"go",t,v)
xlabel("t")
legend("data","v")
```
