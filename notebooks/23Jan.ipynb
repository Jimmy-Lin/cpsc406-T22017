## Application of regularization to signal denoising

```julia
# Generate a "true" signal
n = 300
t = linspace(0,4,n)
x = sin.(t)+t.*(cos.(t).^2)

# add noise
b = x + 0.08*randn(n);
```

```julia
plot(t, b, leg=false)
```

```julia
L = diagm(ones(n)) - diagm(ones(n-1),1)
L = L[1:n-1,:]
@manipulate for λ = 0:1:100
    x = ( eye(n) + λ*L'L ) \ b # solve minimize ||x-b||^2 + λ||Lx||^2
    plot(t, b, leg=false)
    plot!( t, x )
end
```
